```python
pH_cols = [col for col in ITS.columns if 'pH' in col and 'home' not in col]
ITS_pH = ITS[pH_cols]
pH_clean = ITS_pH.dropna()
pH_clean
```

**2.1.3) Time change of pH - Difference of Home and Laboratory**

```python
# read data
pH_cols = [col for col in ITS.columns if 'pH' in col and 'home' not in col]
pH_cols_home = [col for col in ITS.columns if 'pH_home' in col]

ITS_pH = ITS[pH_cols].dropna()
ITS_pH_home = ITS[pH_cols_home].dropna()

pH_means = ITS_pH.mean(axis=0)
pH_std = ITS_pH.std(axis=0)
pH_means_home = ITS_pH_home.mean(axis=0)
pH_std_home = ITS_pH_home.std(axis=0)

# see what the columns are called
print("Lab columns:", pH_means.index.tolist())
print("Home columns:", pH_means_home.index.tolist())
```

```python
# export x-axis
import re

#flexible extraction
def extract_day(col_name):
    # Sucht nach Zahlen nach "day" oder "D"
    match = re.search(r'day(\d+)|D(\d+)', col_name, re.IGNORECASE)
    if match:
        return int(match.group(1) or match.group(2))
    return None

x_lab = [extract_day(col) for col in pH_means.index]
x_home = [extract_day(col) for col in pH_means_home.index]

# Falls None dabei ist, filtern:
data_lab = [(x, y, err) for x, y, err in zip(x_lab, pH_means.values, pH_std.values) if x is not None]
data_home = [(x, y, err) for x, y, err in zip(x_home, pH_means_home.values, pH_std_home.values) if x is not None]

# Plot
plt.figure(figsize=(10, 6))

if data_lab:
    x_lab, y_lab, err_lab = zip(*data_lab)
    plt.errorbar(x=x_lab, y=y_lab, yerr=err_lab, fmt='o-', capsize=5, 
                 ecolor='lightcoral', color='darkred', label='Laboratory')

if data_home:
    x_home, y_home, err_home = zip(*data_home)
    plt.errorbar(x=x_home, y=y_home, yerr=err_home, fmt='s-', capsize=5,
                 ecolor='lightblue', color='darkblue', label='Home')

plt.xlabel('Day', fontsize=16, labelpad=15)
plt.ylabel('pH', fontsize=16, labelpad=15)
plt.title('pH Comparison: Laboratory vs. Home', fontsize=20, pad=15)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
